```r
setwd("/Users/j.amanirad/Desktop/CompositeFace") 
CF_CZ_sum_DP <- read_excel("CF_CZ_sum_DP.xlsx")
CF_CZ_sum_DP

CF_RPPC_sum_DP <- read_excel("CF_RPPC_sum_DP.xlsx")
CF_RPPC_sum_DP

CF_LPPC_sum_DP <- read_excel("CF_LPPC_sum_DP.xlsx")
CF_LPPC_sum_DP
```

```r
CF_CZ_sum_dprime_out <- CF_CZ_sum_DP
CF_RPPC_sum_dprime_out <- CF_RPPC_sum_DP
CF_LPPC_sum_dprime_out <- CF_LPPC_sum_DP
```

```r
CF_CZ_sum_dprime_out$group <- "CZ"
CF_RPPC_sum_dprime_out$group <- "RPPC"
CF_LPPC_sum_dprime_out$group <- "LPPC"
```

```r
CF <- rbind(CF_CZ_sum_dprime_out,CF_RPPC_sum_dprime_out, CF_LPPC_sum_dprime_out)
```

```r
CF
```

```r
CF %>%
  group_by(Cong, Alig, group) %>%
  get_summary_stats(dPrime, type = "mean_sd")
```

```r
CF %>%
  group_by(Cong, Alig, group) %>%
  identify_outliers(dPrime)

bxp <- ggboxplot(
  CF, x = "Alig", y = "dPrime",
  color = "Cong", palette = "jco", 
  facet.by = "group", short.panel.labs = FALSE
) 
bxp
```

```r
dd <- CF %>%
  group_by(Cong, Alig, group) %>%
  identify_outliers(dPrime)

#dd <- dd[which(dd$is.extreme == TRUE),]

#qnorm(0.65) - qnorm(0.85)

CF2<- CF %>% 
  anti_join(dd, by = "dPrime") 

CF2 %>%
  group_by(Cong, Alig) %>%
  identify_outliers(dPrime)

bxp <- ggboxplot(
  CF2, x = "Alig", y = "dPrime",
  color = "Cong", palette = "jco", 
  facet.by = "group", short.panel.labs = FALSE
) 
bxp
```

```r
ggdensity(CF2, x = "dPrime",
   add = "mean", rug = TRUE,
   color = "Cong", fill = "Cong",
   palette = c("#00AFBB", "#E7B800"), facet.by = c("Alig", "group"))
```

```r
CF2 %>%
  group_by(Cong, Alig, group) %>%
  get_summary_stats(dPrime, type = "mean_sd")
```

```r
CF2 %>% group_by(Cong, Alig, group) %>% shapiro_test(dPrime)
```

```r
ggqqplot(CF2, "dPrime", ggtheme = theme_bw()) +
  facet_grid(Cong + group ~ Alig, labeller = "label_both")
```

```r
bxp <- ggboxplot(
  CF2, x = "Alig", y = "dPrime",
  color = "Cong", palette = "jco", 
    #outlier.shape = NA, 
  facet.by = "group", short.panel.labs = FALSE
) 
bxp
```

```r
res.aov <- anova_test(
  data = CF2, dv = dPrime, wid = subj,
  within = c(Cong, Alig, group)
)
get_anova_table(res.aov)
```

```r
two.way <- CF2 %>%
  group_by(group) %>%
  anova_test(dv = dPrime, wid = subj, within = c(Cong,Alig)) %>%
  get_anova_table() %>%
  adjust_pvalue(method = "bonferroni")
two.way
```

```r
two.way2 <- CF2 %>%
  group_by(Cong) %>%
  anova_test(dv = dPrime, wid = subj, within = c(Alig, group)) %>%
  get_anova_table() %>%
  adjust_pvalue(method = "bonferroni")
two.way2
```

```r
two.way3 <- CF2 %>%
  group_by(Alig) %>%
  anova_test(dv = dPrime, wid = subj, within = c(Cong, group)) %>%
  get_anova_table() %>%
  adjust_pvalue(method = "bonferroni")
two.way3
```
