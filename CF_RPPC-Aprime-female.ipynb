```r
for(it in 1:length(CF_RPPC$subj)){
  if(CF_RPPC$subj[it] >= 10 & CF_RPPC$subj[it]<= 27){
    if(CF_RPPC$Key_Resp[it] == "A" & CF_RPPC$Cor_Ans[it] == "A")
      CF_RPPC$SDT[it] <- "Hit"
    else if(CF_RPPC$Key_Resp[it] == "A" & CF_RPPC$Cor_Ans[it] == "L")
      CF_RPPC$SDT[it] <- "FalseAlarm"
    else if(CF_RPPC$Key_Resp[it] == "L" & CF_RPPC$Cor_Ans[it] == "A")
      CF_RPPC$SDT[it] <- "Miss"
    else
      CF_RPPC$SDT[it] <- "CorrReject"
  }
  else{
    if(CF_RPPC$Key_Resp[it] == "L" & CF_RPPC$Cor_Ans[it] == "L")
      CF_RPPC$SDT[it] <- "Hit"
    else if(CF_RPPC$Key_Resp[it] == "L" & CF_RPPC$Cor_Ans[it] == "A")
      CF_RPPC$SDT[it] <- "FalseAlarm"
    else if(CF_RPPC$Key_Resp[it] == "A" & CF_RPPC$Cor_Ans[it] == "L")
      CF_RPPC$SDT[it] <- "Miss"
    else
      CF_RPPC$SDT[it] <- "CorrReject"
  }
}
CF_RPPC
```

included in the analyses.

- binom.test(93,160,p=0.5)  ,  p_value = 0.04777
- binom.test(94,160,p=0.5)  ,  p_value = 0.03247

```r
CF_RPPC_sum <- data.frame(subj = 19,
                       # Cond = NaN,
                        
                        Cong = NaN,
                        Alig = NaN,
                        Acc = NaN,
                        MRT = NaN,
                        SDT = NaN,
                        NumbersOf = NaN,
                        HitRate = NaN,
                        FalseAlarmRate = NaN,
                        dPrime = NaN,
                        bias = NaN)

CF_RPPC_sum2 <- data.frame(subj = NaN,
                      #  Cond = NaN,
                        
                        Cong = NaN,
                        Alig = NaN,
                        Acc = NaN,
                        MRT = NaN,
                        SDT = NaN,
                        NumbersOf = NaN,
                        HitRate = NaN,
                        FalseAlarmRate = NaN,
                        dPrime = NaN,
                        bias = NaN)
```

```r
iter <- 2
for(s in 19:36)
  if(s != 10)
  for(c in c("Congruent", "InCongruent"))
    for(a in c("Alignment", "MissAlignment")){
      for(condition in 1:4){
        CF_RPPC_sum$subj <- s
        CF_RPPC_sum$Cong <- c
          CF_RPPC_sum$Alig <- a
        if(condition == 1){
          CF_RPPC_sum$SDT <- "Hit"
          #CF_CZ_sum$Cond <- "Same"
          CF_RPPC_sum$NumbersOf <- length(CF_RPPC_no_outlier$Acc[CF_RPPC_no_outlier$subj == s & 
                                                               CF_RPPC_no_outlier$Cong == c & 
                                                               CF_RPPC_no_outlier$Alig == a & 
                                                               (CF_RPPC_no_outlier$SDT == "Hit")])
          CF_RPPC_sum$Acc <- mean(CF_RPPC_no_outlier$Acc[CF_RPPC_no_outlier$subj == s & 
                                                        CF_RPPC_no_outlier$Cong == c & 
                                                        CF_RPPC_no_outlier$Alig == a & 
                                                        (CF_RPPC_no_outlier$SDT == "Hit" | CF_RPPC_no_outlier$SDT == "Miss")])
          CF_RPPC_sum$MRT <- mean(CF_RPPC_no_outlier$RT[CF_RPPC_no_outlier$subj == s & 
                                                        CF_RPPC_no_outlier$Cong == c & 
                                                        CF_RPPC_no_outlier$Alig == a & 
                                                        (CF_RPPC_no_outlier$SDT == "Hit")])
          iter <- iter + 1
        }
        else if(condition == 2){
          CF_RPPC_sum$SDT <- "Miss"
       #   CF_CZ_sum$Cond <- "Same"
          CF_RPPC_sum$NumbersOf <- length(CF_RPPC_no_outlier$Acc[CF_RPPC_no_outlier$subj == s & 
                                                               CF_RPPC_no_outlier$Cong == c & 
                                                               CF_RPPC_no_outlier$Alig == a & 
                                                               (CF_RPPC_no_outlier$SDT == "Miss")])
            CF_RPPC_sum$Acc <- NaN
          #CF_CZ_sum$Acc <- mean(CF_CZ_no_outlier$Acc[CF_CZ_no_outlier$subj == s & 
           #                                             CF_CZ_no_outlier$Cong == c & 
           #                                             CF_CZ_no_outlier$Alig == a & 
            #                                            (CF_CZ_no_outlier$SDT == "Miss")])
          CF_RPPC_sum$MRT <- mean(CF_RPPC_no_outlier$RT[CF_RPPC_no_outlier$subj == s & 
                                                      CF_RPPC_no_outlier$Cong == c & 
                                                      CF_RPPC_no_outlier$Alig == a & 
                                                      (CF_RPPC_no_outlier$SDT == "Miss")])
          CF_RPPC_sum2$HitRate[iter - 1] <- CF_RPPC_sum2$NumbersOf[iter -1]/(CF_RPPC_sum2$NumbersOf[iter -1] + CF_RPPC_sum$NumbersOf)
          Hitrate <- CF_RPPC_sum2$HitRate[iter - 1]
          iter <- iter + 1
        }
        else if(condition == 3){
          CF_RPPC_sum$SDT <- "FalseAlarm"
      #    CF_CZ_sum$Cond <- "Different"
          CF_RPPC_sum$NumbersOf <- length(CF_RPPC_no_outlier$Acc[CF_RPPC_no_outlier$subj == s & 
                                                               CF_RPPC_no_outlier$Cong == c & 
                                                               CF_RPPC_no_outlier$Alig == a & 
                                                               (CF_RPPC_no_outlier$SDT == "FalseAlarm")])
            CF_RPPC_sum$Acc <- NaN
            
        #  CF_CZ_sum$Acc <- mean(CF_CZ_no_outlier$Acc[CF_CZ_no_outlier$subj == s & 
         #                                              CF_CZ_no_outlier$Cong == c & 
          #                                             CF_CZ_no_outlier$Alig == a & 
           #                                            (CF_CZ_no_outlier$SDT == "FalseAlarm")])
          CF_RPPC_sum$MRT <- mean(CF_RPPC_no_outlier$RT[CF_RPPC_no_outlier$subj == s & 
                                                      CF_RPPC_no_outlier$Cong == c & 
                                                      CF_RPPC_no_outlier$Alig == a & 
                                                      (CF_RPPC_no_outlier$SDT == "FalseAlarm")])
          iter <- iter + 1
        }
        else if(condition == 4){
          CF_RPPC_sum$SDT <- "CorrReject"
     #     CF_CZ_sum$Cond <- "Different"
          CF_RPPC_sum$NumbersOf <- length(CF_RPPC_no_outlier$Acc[CF_RPPC_no_outlier$subj == s & 
                                                               CF_RPPC_no_outlier$Cong == c & 
                                                               CF_RPPC_no_outlier$Alig == a & 
                                                               (CF_RPPC_no_outlier$SDT == "CorrReject")])
          CF_RPPC_sum$Acc <- mean(CF_RPPC_no_outlier$Acc[CF_RPPC_no_outlier$subj == s & 
                                                       CF_RPPC_no_outlier$Cong == c & 
                                                       CF_RPPC_no_outlier$Alig == a & 
                                                       (CF_RPPC_no_outlier$SDT == "CorrReject" | CF_RPPC_no_outlier$SDT == "FalseAlarm")])
          CF_RPPC_sum$MRT <- mean(CF_RPPC_no_outlier$RT[CF_RPPC_no_outlier$subj == s & 
                                                      CF_RPPC_no_outlier$Cong == c & 
                                                      CF_RPPC_no_outlier$Alig == a & 
                                                      (CF_RPPC_no_outlier$SDT == "CorrReject")])
          CF_RPPC_sum2$FalseAlarmRate[iter - 3] <- CF_RPPC_sum2$NumbersOf[iter -1]/(CF_RPPC_sum2$NumbersOf[iter -1] + CF_RPPC_sum$NumbersOf)
          Falsealarmrate <- CF_RPPC_sum2$FalseAlarmRate[iter - 3]
          iter <- iter + 1
        }
        CF_RPPC_sum2 <- rbind(CF_RPPC_sum2,CF_RPPC_sum)
      }
      le<- length(CF_RPPC_no_outlier$Acc[CF_RPPC_no_outlier$subj == s & 
                                                        CF_RPPC_no_outlier$Cong == c & 
                                                        CF_RPPC_no_outlier$Alig == a ])
      if(Hitrate == 1)
          qnorm_hit <- qnorm((le-0.5)/le)
      else if(Hitrate == 0)
          qnorm_hit <- qnorm((0.5)/le)
      else 
          qnorm_hit <- qnorm(Hitrate)
      if(Falsealarmrate == 1)
          qnorm_fa <- qnorm((le-0.5)/le)
      else if(Falsealarmrate == 0)
          qnorm_fa <- qnorm((0.5)/le)
      else 
          qnorm_fa <- qnorm(Falsealarmrate)
      #CF_CZ_sum2$dPrime[iter - 4] <- pnorm((qnorm_hit - qnorm_fa)/sqrt(2))
      CF_RPPC_sum2$bias[iter - 4] <- -0.5 * (qnorm_hit + qnorm_fa)
      CF_RPPC_sum2$dPrime[iter - 4] <- 0.5 + sign(Hitrate - Falsealarmrate) * 
          (((Hitrate - Falsealarmrate)^2 + abs(Hitrate - Falsealarmrate)) / 
          (4 * max(Hitrate , Falsealarmrate) - 4 * Hitrate * Falsealarmrate))
    }
```

```r
CF_RPPC_sum <- data.frame(subj = 19,
                       # Cond = NaN,
                        Cong = NaN,
                        Alig = NaN,
                        MRT = NaN
                        )

CF_RPPC_sum2 <- data.frame(subj = NaN,
                      #  Cond = NaN,
                        Cong = NaN,
                        Alig = NaN,
                        MRT = NaN
                        )
```

```r
iter <- 2
for(s in 19:36)
  if(s != 10)
  for(c in c("Congruent", "InCongruent"))
    for(a in c("Alignment", "MissAlignment")){
        CF_RPPC_sum$subj <- s
        CF_RPPC_sum$Cong <- c
        CF_RPPC_sum$Alig <- a
        CF_RPPC_sum$MRT <- mean(CF_RPPC_no_outlier$RT[CF_RPPC_no_outlier$subj == s & 
                                                        CF_RPPC_no_outlier$Cong == c & 
                                                        CF_RPPC_no_outlier$Alig == a  
                                                        ])
        iter <- iter + 1
        
        
        CF_RPPC_sum2 <- rbind(CF_RPPC_sum2,CF_RPPC_sum)

    }

```

```r
CF_RPPC_sum_RT <- CF_RPPC_sum2

CF_RPPC_sum_RT <- CF_RPPC_sum_RT[-c(1),]
```
