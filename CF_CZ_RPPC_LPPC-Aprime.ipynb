```r
two.way <- CF %>%
  group_by(group) %>%
  anova_test(dv = dPrime, wid = subj, within = c(Cong,Alig)) %>%
  get_anova_table() %>%
  adjust_pvalue(method = "bonferroni")
two.way
```

```r
two.way2 <- CF %>%
  group_by(Cong) %>%
  anova_test(dv = dPrime, wid = subj, within = c(Alig, group)) %>%
  get_anova_table() %>%
  adjust_pvalue(method = "bonferroni")
two.way2
```
